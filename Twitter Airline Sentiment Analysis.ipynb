```python
x = df.iloc[:, 1]
y = df.iloc[:, 0]
```

```python
import string
from nltk.corpus import stopwords
stops = stopwords.words("english")
stops = stops + list(string.punctuation)
```

```python
from nltk.tokenize import word_tokenize
```

```python
s = "the sky is blue"
word_tokenize(s)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
```

```python
word = "nice"
pos_tag([word])
get_simple_pos(pos_tag([word])[0][1])
```

```python
for i in range(len(x)):
    tweet = x.iloc[i]
    words = word_tokenize(tweet)
    cleaned_words = []
    for word in words:
        if word.lower() not in stops and word.lower() not in airlines:
            cleaned_word = lemmatizer.lemmatize(word, pos = get_simple_pos(pos_tag([word])[0][1]))
            cleaned_words.append(cleaned_word)
    tweet = " ".join(cleaned_words)
    x.iloc[i] = tweet
```

```python
# you may make a set of the airline column and remove the names of the airlines from the tweet to avoid the model from getting cluttered
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size = 0.3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec = CountVectorizer(max_features = 2000, ngram_range = (1, 2))
x_train_features = count_vec.fit_transform(x_train)
x_train_features
```

```python
x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features = count_vec.transform(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(x_train_features, y_train)
y_pred = clf.predict(x_test_features)
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(x_train_features, y_train)
y_pred = clf.predict(x_test_features)
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train_features, y_train)
y_pred = clf.predict(x_test_features)
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(x_train_features.todense(), y_train)
y_pred = clf.predict(x_test_features.todense())
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(x_train_features.todense(), y_train)
y_pred = clf.predict(x_test_features.todense())
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# dividing training set into 3 sets and will then use majority of the prediction
```

```python
x_train_batches = []
y_train_batches = []
N = x_train_features.shape[0]
start = 0
end = int(N/3)
for j in range(3):
    x_train_batches.append(x_train_features[start: end, :])
    y_train_batches.append(y_train[start: end])
    start = end
    end = int(end + N/3)
```

```python
y_pred_batches = []
for j in range(3):
    clf = LogisticRegression()
    clf.fit(x_train_batches[j], y_train_batches[j])
    y_pred_batches.append(clf.predict(x_test_features))
```

```python
y_pred = []
for i in range(len(x_test_features.todense())):
    neg = 0
    pos = 0
    neu = 0
    for j in range(3):
        if y_pred_batches[j][i] == 'negative':
            neg += 1
        elif y_pred_batches[j][i] == 'positive':
            pos += 1
        else:
            neu += 1
    if pos >= neg and pos >= neu:
        y_pred.append('positive')
    elif neg >= pos and neg >= neu:
        y_pred.append('negative')
    else:
        y_pred.append('neutral')
```

```python
print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
