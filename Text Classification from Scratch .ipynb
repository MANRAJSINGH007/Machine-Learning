```python
import sklearn.datasets as skd
categories = ['alt.atheism',
    'comp.graphics',
    'comp.os.ms-windows.misc',
    'comp.sys.ibm.pc.hardware',
    'comp.sys.mac.hardware',
    'comp.windows.x',
    'misc.forsale',
    'rec.autos',
    'rec.motorcycles',
    'rec.sport.baseball',
    'rec.sport.hockey',
    'sci.crypt',
    'sci.electronics',
    'sci.med',
    'sci.space',
    'soc.religion.christian',
    'talk.politics.guns',
    'talk.politics.mideast'
    'talk.politics.misc',
    'talk.religion.misc']
news_train = skd.load_files('C:/Users/lenovo/Desktop/text classification data/20news-bydate-train', categories= categories, encoding= 'ISO-8859-1')
news_test = skd.load_files('C:/Users/lenovo/Desktop/text classification data/20news-bydate-test',categories= categories, encoding= 'ISO-8859-1')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
stop_words.add('')
```

```python
word_freq = {}
for document in news_train.data:
    words = document.split(' ')
    for word in words:
        curr_word = word.lower()
        if (curr_word not in stop_words and curr_word not in word_freq):
            word_freq[curr_word] = 1
        elif(curr_word not in stop_words):
            word_freq[curr_word] += 1
```

```python
import operator
sorted_x = sorted(word_freq.items(), key=operator.itemgetter(1))
```

```python
sorted_x[-24000:]
```

```python
len(sorted_x)
```

```python
sorted_x = sorted_x[-24000:]
```

```python
len(sorted_x)
```

```python
index = 0
word_to_number = {}
for j in sorted_x:
    curr_word = j[0]
    if curr_word not in word_to_number:
        word_to_number[curr_word] = index
        index += 1
```

```python
import numpy as np
vocab = np.zeros((len(news_train.data), len(sorted_x)))
```

```python
print(vocab.shape)
```

```python
len(word_to_number)
```

```python
for i in range(len(news_train.data)):
    document = news_train.data[i]
    words = document.split(' ')
    for word in words:
        curr_word = word.lower()
        if curr_word in word_to_number.keys():
            word_index = word_to_number[curr_word]
            vocab[i, word_index] += 1
```

```python
set(news_train.target)
```

```python
result = {}
for curr_class in set(news_train.target):
    result[curr_class] = {}
```

```python
for i in range(len(news_train.data)):
    curr_class_name = news_train.target[i]
    if "total_count" in result[curr_class_name].keys():
        result[curr_class_name]["total_count"] += 1
    else:
        result[curr_class_name]["total_count"] = 1
```

```python
vocab.shape
```

```python
for i in range(vocab.shape[0]):
    for j in range(vocab.shape[1]):
        curr_word_index = j
        curr_class_index = news_train.target[i]
        if curr_word_index not in result[curr_class_index].keys():
            result[curr_class_index][curr_word_index] = vocab[i, j]
        else:
            result[curr_class_index][curr_word_index] += vocab[i, j]
```

```python
for i in range(vocab.shape[0]):
    s = 0
    for j in range(vocab.shape[1]):
        s += vocab[i, j]
    curr_class_index = news_train.target[i]
    if "total_words" not in result[curr_class_index].keys():
        result[curr_class_index]["total_words"] = s
    else:
        result[curr_class_index]["total_words"] += s
```

```python
predicted = []
N = len(news_train.data)
for i in range(len(news_test.data)):
    document = news_test.data[i]
    words = document.split(' ')
    best_class = -10000
    best_probability = -10000
    for k in range(len(news_train.target_names)):
        curr_class_index = k
        flag = 0
        output = np.log(result[curr_class_index]["total_count"]) - np.log(N)
        for w in words:
            word = w.lower()
            if word in word_to_number:
                word_index = word_to_number[word]
                flag = 1
                if word_index in result[curr_class_index].keys():
                    count1 = result[curr_class_index][word_index]
                    count2 = result[curr_class_index]["total_words"]
                    output += np.log(count1 + 1) - np.log(count2 + vocab.shape[1])
                else:
                    output -= np.log(result[curr_class_index]["total_words"] + vocab.shape[1])
        if(flag == 0):
            output = -10000
        if(flag == 1 and output > best_probability):
            best_probability = output
            best_class = curr_class_index
    predicted.append(best_class)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
print("Accuracy: ", accuracy_score(news_test.target, predicted))
print(metrics.classification_report(news_test.target, predicted, target_names = news_test.target_names))
print("Confusion Matrix: "+"\n",metrics.confusion_matrix(news_test.target, predicted))
```
