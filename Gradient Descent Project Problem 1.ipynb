```python
def gd(data, learning_rate, number_of_iterations):
    slopes = np.zeros(data.shape[1])
    new_slopes = np.zeros(data.shape[1])
    for i in range(number_of_iterations):
        for j in range(len(slopes)):
            new_slopes[j] = slopes[j] - learning_rate*step_gradient(data,slopes,j)
        slopes = new_slopes
        print(i,"th iteration with cost ",cost(data,slopes))
    return slopes
```

```python
def step_gradient(data, slopes, j):
    M = len(data)
    N = len(slopes)
    if j==(N-1):
        change = 0
        for i in range(M):
            pred = 0
            for k in range(N-1):
                pred += slopes[k]*data.iloc[i, k]
            pred += slopes[N-1]
            change += (-2/M) * (data.iloc[i, N-1] - pred)*1
        return change
    else:
        change = 0
        for i in range(M):
            pred = 0
            for k in range(N-1):
                pred += slopes[k]*data.iloc[i, k]
            pred += slopes[N-1]
            change += (-2/M) * (data.iloc[i, N-1] - pred)*data.iloc[i, j]
        return change
```

```python
def cost(data, slopes):
    N = len(slopes)
    M = len(data)
    total_cost = 0
    for i in range(M):
        pred = 0
        for j in range(N-1):
            pred += slopes[j]* data.iloc[i,j]
        pred += slopes[N-1]
        total_cost += (data.iloc[i,N-1] - pred)**2
    return total_cost
```

```python
def run(data):
    learning_rate = 0.1
    number_of_iterations = 100
    return gd(data, learning_rate, number_of_iterations)
```

```python
slopes = run(boston)
```

```python
for i in range(len(slopes)):
    print("Slope for ",i,"th feature is ",slopes[i])
```

```python
def predict(x, slopes):
    pred = np.zeros(len(x))
    for i in range(len(x)):
        this_pred = slopes[len(slopes)-1]
        for j in range(len(slopes)-1):
            this_pred += slopes[j]* x.iloc[i,j]
        pred[i] = this_pred
    return pred
```

```python
y_pred = predict(pd.read_csv("test_boston_x_test.csv"),slopes)
```
